### EXPLORATORY DATA ANALYSIS (EDA) of IPL DATASET ###

```python
# importing necessary libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
# Reading the csv files 
mdf = pd.read_csv('matches.csv')
ddf = pd.read_csv('deliveries.csv')
```

```python
# merginf 2 dataset 
df = pd.merge(ddf, mdf, left_on='match_id', right_on='id')
df.head()
```

```python
# No.of matches played in each season
plt.figure(figsize=(16,8))
sns.countplot(x='season', data=mdf)
plt.title('NO.of Matches PLayed in each season', fontsize=15)
plt.xlabel('Season', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.ylabel('No.of Matches', fontsize=15)
```

```python
# Teams played in each season 
plt.figure(figsize=(16,8))
season_team = mdf.groupby(['season'], as_index=False)['team1'].nunique() 
#.sum().sort_values(by="team1", ascending=False) is not used here it is mostly used for numerical value 
#.nunique is used for categorical value like team 1 which is used unique  
sns.barplot(x='season', y='team1', data=season_team)

plt.title("Team played in eavh Season", fontsize=15)
plt.xlabel('Season', fontsize=15)
plt.ylabel("Number of team", fontsize=15)
plt.show()
```

```python
# Venue where most maches occured 
plt.figure(figsize=(16,8))
sns.countplot(x='venue', data=mdf, order=df['venue'].value_counts().index)

plt.xlabel('Venue', fontsize=15)
plt.ylabel('Number of matches', fontsize=15)
plt.xticks(rotation=90, fontsize=15)
plt.show()
```

```python
# Most winning Teams
winning_team = mdf[['season','winner']]
winner_team = {}
for i in sorted(winning_team.season.unique()):
    winner_team[i] = winning_team[winning_team.season == i]['winner'].tail(1).values[0]
    
winner_IPL = pd.Series(winner_team)
winner_IPL = pd.DataFrame(winner_IPL, columns=['Yearly Winning Team'])
winner_IPL
```

```python
# Outstanding team which has won most of the IPL
plt.figure(figsize=(16,8))
sns.countplot(x='Yearly Winning Team', data=winner_IPL)

plt.show()
```

```python
#TOSS DECISION
plt.figure(figsize=(16,8))
(mdf['toss_decision'].value_counts()).plot(kind='pie',startangle=210,autopct = '%2.3f%%',shadow=4,explode=(0,0.1),colors=['r','orange'], fontsize=15)
plt.title('Decision taken by captain after winning toss', fontsize=15)
```

```python
# Toss decision Affect winning 
mdf['tosswin_matchwin'] = np.where((mdf.toss_winner == mdf.winner),'Yes','No')

plt.figure(figsize=(16,8))
sns.countplot(x='tosswin_matchwin', data=mdf, hue='toss_decision', palette='viridis')
plt.title("Toss decision affect match result", fontsize=15)
plt.xlabel('Winning toss and winning match', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(rotation=0, fontsize=15)
```

```python
#individual team toss decision after winning toss
plt.figure(figsize=(16,8))
sns.countplot(x='toss_winner', data=mdf, hue='toss_decision', palette='rocket')
plt.title('Individual team toss decision after winning toss', fontsize=15)
plt.xlabel('Teams', fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.ylabel('Frequency', fontsize=15)
```

```python
# Top player with most man of the match
plt.figure(figsize=(16,8))
sns.countplot(x='player_of_match', data=mdf, order=df['player_of_match'].value_counts().head(15).index, palette='muted')
plt.title('Top players with most man of the match ', fontsize=15)
plt.xlabel('Players', fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.ylabel('Frequency', fontsize=15)
```

```python
# TOP RUN SCORERS OF IPL
#using merged dataset (mathces and deliveries)
run = df.groupby(['batsman'], as_index=False)['batsman_runs'].sum().sort_values(by="batsman_runs", ascending=False).head(10)
sns.barplot(x='batsman', y='batsman_runs', data=run)
plt.title('Top Run Getters of IPL', fontsize=15)
plt.xlabel('Batsmen', fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.ylabel('Total runs scored', fontsize=15)
```

```python
# Most consistent batsman of IPL
cons_batsman = df[df.batsman.isin(['V Kohli', 'SK Raina', 'RG Sharma', 'DA Warner', 'S Dhawan','CH Gayle','MS Dhoni',
                                         'RV Uthappa', 'AB de Villers', 'G Gambhir'])]
[['batsman', 'season', 'total_runs']]
plt.figure(figsize=(20,10))
cons = cons_batsman.groupby(['season','batsman'], as_index=False)['total_runs'].sum().sort_values(by='total_runs', ascending=False)
sns.boxplot(x='batsman', y='total_runs', data=cons)

plt.show()
```

```python
# Top wicket takers of IPL
bowler = df.groupby(['bowler'], as_index=False)['player_dismissed'].count().sort_values(by='player_dismissed', ascending=False).head(10)
plt.figure(figsize=(20,7))
sns.barplot(x='bowler', y='player_dismissed', data=bowler)
```

```python
# TYPES OF DISMISSAL IN ALL SEASONS OF IPL
plt.figure(figsize=(16,8))
dismissal_type = ddf.dismissal_kind.value_counts()
sns.barplot(x = dismissal_type.index, y = dismissal_type,palette="colorblind")
plt.title('Types of Dismissal in all seasons of IPL', fontsize=15)
plt.xlabel('Types of Dismissal', fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.show()
```

```python
#we will consider batsmen, which played most of the seasons
no_of_balls = pd.DataFrame(df.groupby('batsman')['ball'].count())
runs = pd.DataFrame(df.groupby('batsman')['batsman_runs'].sum())
seasons = pd.DataFrame(df.groupby('batsman')['season'].nunique())

strike_rate = pd.DataFrame({'balls':no_of_balls['ball'], 'run':runs['batsman_runs'], 'season':seasons['season']})
strike_rate.reset_index(inplace = True)
strike_rate['strike_rate'] = (strike_rate['run']/strike_rate['balls'])*100
highest_strike_rate = strike_rate[strike_rate.season.isin([9,10,11])][['season','batsman','strike_rate']].sort_values(by='strike_rate',ascending = False)
highest_strike_rate.head(10)
```
